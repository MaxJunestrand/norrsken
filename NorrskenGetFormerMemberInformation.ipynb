```python
import pandas as pd

df = pd.read_excel('Former House Members.xlsx', usecols="A, H, I, N") # can also index sheet by name or fetch all sheets
df["left_norrsken_nr_hires"] = 0
df["left_norrsken_revenue"] = 0
df["left_norrsken_ebitda"] = 0
df["left_norrsken_solidity"] = 0
df["last_nr_hires"] = 0
df["last_revenue"] = 0
df["last_ebitda"] = 0
df["last_solidity"] = 0

df.head()
```

```python
name = 'Översikt - Namn'
revenue = 'Resultaträkning (tkr) - Nettoomsättning'
ebitda = 'Nycketal - Rörelseresultat, EBITDA'
hires = 'Nycketal - Antal anställda'
solidity = 'Nycketal - Soliditet'
company = Company("559093-0334")

def getDataLeavingNorrsken(last_year, column_name, values):
    # Error check
    last_parameter = "null"
    if len(values) > 0:
        # Get value if year exists
        for value in values:
            if last_year == int(value[0][0:4]):
                last_parameter = value[1]
                
        if last_parameter == "null":         
            # If year didn't exist, take closest one        
            if last_year > int(values[0][0][0:4]):
                last_parameter = values[0][1]
            else:
                last_parameter = values[-1][1]
    df[column_name][i] = last_parameter

def getCurrentData(column_name, values):
    # Error check
    last_parameter = "null"
    if len(values) > 0:          
        last_parameter = values[0][1]
    df[column_name][i] = last_parameter    
    
for i in df.index:
    time.sleep(5)
    # Get company information from allbolag
    org_nr = df['Reg Number'][i]
    # Check so organisation number is know.
    if org_nr == "x":
        df["left_norrsken_nr_hires"][i] = "null"
        continue
    # Get all company info
    
    company = Company(df['Reg Number'][i].strip())
    
    # Get company information from year they left norrsken
    # Last year they were member at Norrsken
    last_year_at_norrksen = int(df['End Year'][i])
    getDataLeavingNorrsken(last_year_at_norrksen, "left_norrsken_revenue", company.data[revenue])
    getDataLeavingNorrsken(last_year_at_norrksen, "left_norrsken_nr_hires", company.data[hires])
    getDataLeavingNorrsken(last_year_at_norrksen, "left_norrsken_ebitda", company.data[ebitda])
    getDataLeavingNorrsken(last_year_at_norrksen, "left_norrsken_solidity", company.data[solidity])
    
    getCurrentData("last_revenue", company.data[revenue])
    getCurrentData("last_nr_hires", company.data[hires])
    getCurrentData("last_ebitda", company.data[ebitda])
    getCurrentData("last_solidity", company.data[solidity])


df.head()
```
