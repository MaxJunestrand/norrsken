```python
import pandas as pd

df = pd.read_excel('Former House Members.xlsx', usecols="A, H, I, N") # can also index sheet by name or fetch all sheets
df["left_norrsken_nr_hires"] = 0
df.head()
```

```python
name = 'Översikt - Namn'
revenue = 'Resultaträkning (tkr) - Nettoomsättning'
ebitda = 'Nycketal - Rörelseresultat, EBITDA'
hires = 'Nycketal - Antal anställda'

company = Company("559093-0334")

for i in df.index:
    # Get company information from allbolag
    org_nr = df['Reg Number'][i]
    print(org_nr)
    if org_nr == "x":
        df["left_norrsken_nr_hires"][i] = "null"
        continue
    
    company = Company(df['Reg Number'][i])
    company_revenue = company.data[revenue]
    company_hires = company.data[ebitda]
    company_hires = company.data[hires]
    
    
    # Get company information from year they left norrsken
    # Last year they were member at Norrsken
    last_year_at_norrksen = df['End Year'][i] #2019

    # Hires
    last_nr_hires_at_norrsken = "null"
    if len(company_hires) > 1:
        last_nr_hires_at_norrsken = company_hires[0][1]
        for nr_of_hires in company_hires:
            if last_year_at_norrksen == nr_of_hires[0]:
                last_nr_hires_at_norrsken = nr_of_hires[0]
    print(last_nr_hires_at_norrsken)
    df["left_norrsken_nr_hires"][i] = last_nr_hires_at_norrsken
    
    
    # Hires
    last_nr_hires_at_norrsken = "null"
    if len(company_hires) > 1:
        last_nr_hires_at_norrsken = company_hires[0][1]
        for nr_of_hires in company_hires:
            if last_year_at_norrksen == nr_of_hires[0]:
                last_nr_hires_at_norrsken = nr_of_hires[0]
    print(last_nr_hires_at_norrsken)
    df["left_norrsken_nr_hires"][i] = last_nr_hires_at_norrsken
    
    
    
    
    
    if i > 10:
        break
df.head()
```
